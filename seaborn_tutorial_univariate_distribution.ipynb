About displot :
>> Histogram          by Default
>> Inserts a KernelDE by default

you can add bins by    ( x , bin = )

```python
sns.distplot( iris['SepalLengthCm'])
```

You can rmove the kde 

```python
sns.distplot( iris['SepalLengthCm'] , kde = False)
```

```python
sns.distplot( iris['SepalLengthCm'] , kde = False , rug = True)
```

you can also just plot kde without the histogram . !!!

The plot is normalised so that the area adds to 1

```python
sns.distplot( iris['SepalLengthCm'] , hist = False)
```

Using kdeplot function ::


```python
sns.kdeplot( iris['SepalLengthCm'] , shade = True )
```

using plt.legend() we can draw multiple graphs in one

Note : without using this plt.plot nothing seems to happen wrong @!!!!!!

```python
sns.distplot( iris['SepalLengthCm'] , label = 'Sl')
sns.distplot( iris['SepalWidthCm'] , label = 'sw')

# can be removed
plt.legend()
```



you can also check how your histogram fits over some standard function 

first import stats from scipy

then do fit = stats.norm(or gamma ) or anything else

```python
from scipy import stats
sns.distplot( iris['SepalLengthCm'] , label = 'Sl', kde = False , fit = stats.norm)
```
