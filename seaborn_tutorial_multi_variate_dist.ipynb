Using sns.jointplot 

```python
sns.jointplot( iris['SepalLengthCm'] , iris['SepalWidthCm'])
```

Warning : i don't know much about this type of plot'

You can draw Hexbin plots which is analog of histogram for bivariate data.


```python
with sns.axes_style('white'):
    sns.jointplot( iris['SepalLengthCm'] , iris['SepalWidthCm'] , kind = 'hex' , color = 'k')
```

```python
Using Kernel Density Estimation :
    
    this is plots like a contour plot . 
```

```python
sns.jointplot( iris['SepalLengthCm'] , iris['SepalWidthCm'] , kind = 'kde' )
```
